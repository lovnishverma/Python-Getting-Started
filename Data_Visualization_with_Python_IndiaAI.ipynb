# **Matplotlib**

Matplotlib is the most widely used data visualization library in Python. It provides flexible tools to create static, animated, and interactive 2D and 3D graphs. Matplotlib is seamlessly integrated with data manipulation libraries like NumPy and Pandas, making it a standard for data science and research.



```
!pip install matplotlib
```



````python
import matplotlib.pyplot as plt
````

````python
df
````

````python
# Line Plot

plt.plot(df['Age'], df['Score'], marker='o', linestyle='--', color='black', label="Line 1")
plt.title('Age vs Score')
plt.xlabel('Age')
plt.ylabel('Score')
plt.grid()
plt.legend()
plt.show()
````

````python
# Bar Chart
name = df['Name']
age = df['Age']

plt.bar(name,age, color=['green','blue','pink','black','orange','red'])
plt.title('Name Vs Age')
plt.xlabel('Name')
plt.ylabel('Age')
plt.show()
````

````python
# Scatter Plot

plt.scatter(df['Age'], df['Score'], color='r')
plt.title('Age vs Score')
plt.xlabel('Age')
plt.ylabel('Score')
# plt.grid()
plt.savefig('merahdscatter.png', dpi=1200) # always savefig() befor show()
plt.show()
````

````python
import seaborn as sns
````

````python
df
````

````python
# Scatter plot using seaborn

# sns.set(style='darkgrid')

# Set theme
sns.set_theme(style="whitegrid")

# Create scatter plot

plt.figure(figsize=(6,3))

sns.scatterplot(x='Age', y='Score', data=df)
plt.title('Age vs Score')
plt.xlabel('Age')
plt.ylabel('Score')
plt.show()
````

````python
# pie chart

plt.pie(df['Score'], labels=df['Name'], autopct='%1.1f%%')
plt.title('Score Distribution')
plt.show()
````

````python
# seaborn
sns.barplot(x='Name', y='Score', data=df)
sns.set_palette("dark")
plt.title('Name vs Score')
plt.xlabel('Name')
plt.ylabel('Score')
plt.show()
````

````python
df.describe()
````

````python
# Box plot
# set Palette
# muted, pastel, bright, dark, or colorblind

sns.set_palette("pastel")

plt.boxplot(df['Age'])
plt.title('Box Plot')
plt.show()
````

### Popular Style Options (`style=`)

```python
sns.set_theme(style="darkgrid")   # Default seaborn look
sns.set_theme(style="whitegrid")  # White background + grid
sns.set_theme(style="dark")       # Dark background, no grid
sns.set_theme(style="white")      # White background, no grid
sns.set_theme(style="ticks")      # White background + axis ticks
```

### Popular Palette Options (`palette=`)

#### Qualitative / Categorical

```python
sns.set_palette("deep")
sns.set_palette("muted")
sns.set_palette("pastel")
sns.set_palette("bright")
sns.set_palette("dark")
sns.set_palette("colorblind")
sns.set_palette("Set1")
sns.set_palette("Set2")
```

#### Sequential / Continuous

```python
sns.set_palette("Blues")
sns.set_palette("Greens")
sns.set_palette("Reds")
sns.set_palette("Purples")
sns.set_palette("rocket")
sns.set_palette("mako")
sns.set_palette("flare")
sns.set_palette("crest")
```

### Style + Palette Together

```python
sns.set_theme(
    style="whitegrid",
    palette="deep"
)
```

### Temporary Styling Using Context Managers

```python
with sns.axes_style("whitegrid"):
    with sns.color_palette("pastel"):
        sns.barplot(data=df, x="Category", y="Value")
```

or

```python
with sns.plotting_context("talk"):
    sns.lineplot(data=df, x="Date", y="Sales")
```

### Common Professional Choices

**Presentation-ready:**

```python
sns.set_theme(style="whitegrid", palette="deep")
```

**Modern dashboard:**

```python
sns.set_theme(style="ticks", palette="colorblind")
```

**Research paper:**

```python
sns.set_theme(style="white")
```

**Dark theme:**

```python
sns.set_theme(style="dark", palette="bright")
```

**My most-used combination for ML/Data Science projects:**

```python
sns.set_theme(
    style="whitegrid",
    palette="deep",
    context="notebook"
)
```

It produces clean, professional plots that work well in reports, notebooks, and presentations.
