```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head(10)
```

# **Line Plot**

```python
# plot a line plot for age and marks cloumn
plt.plot(df["Age"], df["Marks"], marker='s', linestyle='--', color='black')
plt.xlabel("Age")
plt.ylabel("Marks")
plt.grid()
plt.savefig("myplot.png", dpi=800) # always save before show()
plt.show()
```

### ðŸ“Œ Common Marker Codes (for reference)

| Marker | Meaning       |
| ------ | ------------- |
| `'o'`  | Circle        |
| `'s'`  | Square        |
| `'^'`  | Triangle up   |
| `'v'`  | Triangle down |
| `'D'`  | Diamond       |
| `'x'`  | X             |
| `'+'`  | Plus          |





### ðŸ“Œ Valid `linestyle` values

| Value    | Meaning       |
| -------- | ------------- |
| `'-'`    | Solid line    |
| `'--'`   | Dashed line   |
| `'-.'`   | Dash-dot line |
| `':'`    | Dotted line   |
| `'None'` | No line       |



# Bar Chart

```python
# plot a Bar Chart fir name and age cloumn
plt.bar(df["Name"], df["Age"], color=['r','b'])
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Bar Chart")
plt.grid()
plt.show()
```

```python
# plot a horizontol bar chart
plt.barh(df["Name"], df["Age"], color=['r','b'])
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Bar Chart")
plt.grid()
plt.show()
```
