```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
```

```python
# 1. Load Iris dataset from URL
iris=pd.read_csv("https://raw.githubusercontent.com/lovnishverma/datasets/refs/heads/main/iriswithheaders.csv")
iris.head()
```

```python
x = iris.iloc[:,:-1]
y = iris.iloc[:,-1]
```

```python
#spliting data in test and train
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#evaluate model performance
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy * 100 , "%")
```

```python
#evaluate model performance
model = SVC()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy * 100 , "%")
```

```python
#evaluate model performance
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy * 100 , "%")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)  # scale entire x before splitting

# split scaled

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import cross_val_score

for k in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_scaled, y, cv=5)
    print(f"k={k} -> Mean Accuracy: {scores.mean():.4f}")
```

```python
#evaluate model performance
model = KNeighborsClassifier(n_neighbors=8)  # You can tune k
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy * 100 , "%")
```

```python
# printing classification report
print(classification_report(y_test, y_pred))
```

# Saving Model Using joblib (recommended)

```python
#save model using joblib
import joblib
joblib.dump(model, "iris_model.joblib")
```

```python
#predicting using saved model
model = joblib.load("iris_model.joblib")
prediction = model.predict([[5.7, 3.2, 5.2, 1.9]])
print(prediction[0])
```

"It is preferable to use joblib (dump & load), which is more efficient on objects that carry large numpy arrays."
— [Scikit-learn documentation](https://scikit-learn.org/stable/model_persistence.html)

#or you can use pickle

```python
#Save model using pickle
import pickle
with open("iris_model.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
#predicting using model saved using pickle
model = pickle.load(open("iris_model.pkl", "rb"))
prediction = model.predict([[5.7, 3.2, 5.2, 1.9]])
print(prediction[0])
```
