# Conditional Statements

```python
marks = int(input("Enter your marks: "))

if marks >= 40:
  print("pass")
else:
  print("fail")
```

# Control flow using If- Elif Else

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
  print("A GRADE")
elif marks >= 80:
  print("B GRADE")
elif marks >= 70:
  print("C GRADE")
elif marks >= 60:
  print("D GRADE")
else:
  print("FAIL")
```

# Loops

```python
states= ['assam', 'bihar', 'chandigarh', 'delhi']

# looping through list

for n in states:
  print(n)
```

```python
for i in range(5):
  print(i)
```

```python
for i in range(1, 11):
  print(i, end = " ")
```

# While Loop

```python
count = 0

while count < 5:
  print(count)
  count += 1
```

**String operations**

```python
string = "python programming"

#string methods
print(string.upper())
print(string.lower())
print(string.title()) # Converts the first letter of every word to uppercase.
print(string.capitalize()) # Makes only the first character uppercase.

print(len(string))

#access characters in string
print(string[0])
print(string[1])
print(string[2])
print(string[3])
print(string[4])
print(string[5])

#replace part of the string
print(string.replace("thon", "gon"))
```
