#A Simple Guide to Outliers (With Pizza Example!)
Perfect for Beginners! üçï

**1. What is an Outlier?**

An outlier is a data point that is very different from the rest.

Example:

Imagine you asked 5 friends:

"How many slices of pizza can you eat in one sitting?"

Friend	Slices Eaten

Alice	4

Bob	3

Carol	5

Dave	4

Emma	12

**üî¥ Emma is an outlier!** (Most people eat 3-5 slices, but Emma eats 12!)

**2. How to Detect Outliers?**

Method 1: The "Eye Test" (Just Look at the Data!)

Plot the data and see if any points look too high or too low.

```python
import matplotlib.pyplot as plt
# import seaborn as sns

slices = [4, 3, 5, 4, 12]
friends = ["Alice", "Bob", "Carol", "Dave", "Emma"]

plt.bar(friends, slices, color='skyblue')
plt.title("Pizza Slices Eaten by Friends")
plt.ylabel("Slices")
plt.show()
```

üìå Emma‚Äôs bar is way taller! ‚Üí Outlier!



---



Method 2 : Box plots are one of the most common and effective methods to detect outliers in a dataset ‚Äî especially in univariate (single-variable) analysis.

Let's use our pizza example to create simple box plots that show outliers clearly.

```python
# Create a box plot
plt.figure(figsize=(8, 5))
plt.boxplot(slices, patch_artist=True,
           boxprops=dict(facecolor='lightblue'),
           medianprops=dict(color='red'))
plt.title("Pizza Slices Eaten - Box Plot")
plt.ylabel("Number of Slices")
plt.xticks([1], ["Pizza Consumption"])  # Just one box
plt.show()
```

**Understanding the Box Plot**

**The Box** shows where most people fall (25th to 75th percentile)

**The Red Line** is the median (middle value)

**The Whiskers** show the "normal" range (up to 1.5√óIQR)

**Emma's 12 slices** appears as a dot ‚Üí **outlier!**



---



Method 3: Using Scatter Plot

```python
plt.scatter(slices, friends, color='orange')  # x = slices, y = friends
plt.title("Pizza Slices Eaten by Friends")
plt.xlabel("Slices Eaten")
plt.ylabel("Friends")
plt.grid(True)
plt.show()
```

**Method 4: The IQR Rule (Simple Math)**

IQR (Interquartile Range) helps find "normal" ranges.

**Step 1: Sort the Data**

[3, 4, 4, 5, 12]

**Step 2: Find Q1 (25%) and Q3 (75%)**

Q1 (25%) = 4

Q3 (75%) = 5

**Step 3: Calculate IQR**

IQR = Q3 - Q1 = 5 - 4 = 1

**Step 4: Find the "Normal" Range**

Lower Bound = Q1 - 1.5 √ó IQR = 4 - 1.5 = 2.5

Upper Bound = Q3 + 1.5 √ó IQR = 5 + 1.5 = 6.5

**‚úÖ Normal Range: 2.5 to 6.5 slices**

**‚ùå Emma‚Äôs slices (12) > 6.5 ‚Üí Outlier!**



---



**3. How to Handle Outliers?**

**Option 1: Remove It (If It‚Äôs a Mistake)**

```python
clean_data = [x for x in slices if x <= 6.5]
print(clean_data)  # [4, 3, 5, 4]
```

**Option 2: Cap It (If It‚Äôs Extreme but Real)**

```python
capped_data = [min(x, 6.5) for x in slices]
print(capped_data)  # [4, 3, 5, 4, 6.5]
```

**Option 3: Keep It (If It‚Äôs Important)**

Maybe Emma is a competitive eater! üèÜ

Don‚Äôt remove if it‚Äôs a real (but rare) case.

**4. Let‚Äôs Try Another Example!**

**Data: Daily Coffee Cups Drunk by Office Workers**

[2, 3, 1, 2, 8, 3, 2, 1, 0, 15]

üîé Outlier?

Most drink 0-3 cups.

"15" is an outlier!

**Fixing It:**

```python
coffee = [2, 3, 1, 2, 8, 3, 2, 1, 0, 15]

# Remove outliers (keep ‚â§ 6)
clean_coffee = [x for x in coffee if x <= 6]

# Or cap at 6
capped_coffee = [min(x, 6) for x in coffee]
```
