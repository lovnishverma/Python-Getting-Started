2. Variables and Data Types
Python is dynamically typed, so you don’t need to declare variable types explicitly.

```python
# Variables and data types
name = "Lovnish"         # String
age = 25              # Integer
height = 5.8          # Float
is_student = True     # Boolean

print(name, "is", age,"years old his height is", height, "and it is", is_student, "he/she is a student")
```

3. Input from User

```python
# Input example
name = input("Enter your name: ")
print("Hello, " + name + " 😀")
```

4. Conditional Statements

```python
# Example of if-else
age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult.")
else:
    print("You are not an adult.")
```

5. **Loops**
For Loop:

```python
# Example of a for loop
for i in range(5):  # Loops from 0 to 4
    print(i)
```

While Loop:

```python
# Example of a while loop
count = 0
while count < 5:
    print(count)
    count += 1
```

6. Functions

```python
# Example of a function
def greet(name):
    print(f"Hello, {name}!")

greet("Lovnish")
```

Type Casting:
You can convert data types using:

int(), float(), str(), bool()

```python
# Converting data types
age = "25"       # String
age = int(age)   # Convert to integer
print(age + 5)   # Now works as an integer
```

2. String Operations:

```python
message = "Hello, Python!"

# String methods
print(message.upper())         # Converts to uppercase
print(message.lower())         # Converts to lowercase
print(message.replace("Python", "World"))  # Replace part of the string
print(len(message))            # Length of the string

# Access characters in a string
print(message[0])              # First character
print(message[-1])             # Last character
print(message[7:13])           # Slicing: "Python"
```

3. Lists, Tuples, and Dictionaries
Lists (Mutable - Can Change)

```python
# A list of items
fruits = ["apple", "banana", "cherry"]

# Accessing items
print(fruits[0])          # Access first element
print(fruits[-1])         # Access last element

# Adding items
fruits.append("mango")    # Add at the end
print(fruits)

# Removing items
fruits.remove("banana")
print(fruits)

# Looping through a list
for fruit in fruits:
    print(fruit)
```

Tuples (Immutable - Cannot Change)

```python
# Tuples are like lists but cannot be modified
coordinates = (10, 20)
print(coordinates[0])    # Access element
```

Dictionaries (Key-Value Pairs)

```python
# A dictionary stores data in key-value pairs
person = {"name": "Lovnish", "age": 25, "city": "Ropar"}

# Accessing values
print(person["name"])      # Get value by key

# Adding or modifying values
person["age"] = 26         # Update age
person["job"] = "Project Engineer" # Add a new key-value pair

print(person)

# Looping through a dictionary
for key, value in person.items():
    print(key, ":", value)
```

4. Control Flow in Detail
If-Elif-Else

```python
# Nested conditions
score = int(input("Enter your score: "))

if score >= 90:
    print("Grade: A")
elif score >= 75:
    print("Grade: B")
elif score >= 50:
    print("Grade: C")
else:
    print("Grade: F")
```

For Loops (with Range)

```python
# Loop through a range of numbers
for i in range(1, 11):  # From 1 to 10
    print(i, end=" ")   # Print on the same line
```

While Loop (with a Condition)

```python
# Continue looping until condition is met
count = 0
while count < 5:
    print("Count:", count)
    count += 1
```

5. Functions Basics

```python
def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")

greet("Lovnish", 25)
```

Return Values

```python
def square(num):
    return num * num

result = square(5)
print("Square:", result)
```

Default Arguments

```python
def greet(name="User"):
    print(f"Hello, {name}!")

greet()                 # Uses the default value
greet("Lovnish")            # Overrides the default
```

6. Working with Files
Reading a File


```python
# Open a file in write mode
with open("example.txt", "w") as file:
    file.write("This is a test file.")
```

Working with Files
Reading a File

```python
# Open a file in read mode
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```
