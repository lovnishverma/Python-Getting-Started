```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
```

```python
# 1. Load Iris dataset from URL
iris=pd.read_csv("https://raw.githubusercontent.com/lovnishverma/datasets/refs/heads/main/iriswithheaders.csv")
```

```python
# Box plot for all features
iris.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
print(iris.describe())
plt.suptitle("Box Plots of Iris Dataset Features")
plt.show()
```
