## **1. Regression**

* **Definition**: Predicts a **continuous numeric value**.
* **Output**: Real numbers (e.g., 3.14, 200, 50000).
* **Examples**:

  * Predicting **house price** from features like size, location, and rooms.
  * Predicting **temperature** tomorrow.
  * Predicting **sales revenue** next month.

**Visualization Idea**: A best-fit line predicting values.

```python

import numpy as np

X = np.array([[1], [2], [3], [4], [5]])   # feature
y = np.array([2, 4, 6, 8, 10])            # target (continuous)

model = LinearRegression()
model.fit(X, y)

print(model.predict([[6]]))  # Predict value for feature=6
```

**Output:** `[12.]`

## **2. Classification**

* **Definition**: Predicts a **category/class**.
* **Output**: Discrete values (labels).
* **Examples**:

  * Predicting if a person has **diabetes** → Yes (1) / No (0).
  * Predicting if an email is **spam** or **not spam**.
  * Predicting the **digit** (0–9) in handwritten images (MNIST dataset).

 **Visualization Idea**: Decision boundaries separating classes.

## **3. Encoding Categorical Data**

(Machine learning models only understand numerical data, so categorical features must be converted into numbers.)

There are two common encoding techniques:

1. **Label Encoding**

   * Example: `Male / Female / Trans` → `0 / 1 / 2`
   * Used when categories are ordinal (have a natural order).

2. **One-Hot Encoding**

   * Example:

     * Male → `Is_Male = 1, Is_Female = 0, Is_Trans = 0`
     * Female → `Is_Male = 0, Is_Female = 1, Is_Trans = 0`
     * Trans → `Is_Male = 0, Is_Female = 0, Is_Trans = 1`
   * Used when categories are nominal (no order).

## **4. Feature Scaling**

Scaling ensures that features are on a similar scale, which improves convergence and accuracy in ML models.

1. **Min-Max Scaler**

   * Scales values to a range of `[0, 1]`.
   * Formula:

     $$
     X' = \frac{X - X_{min}}{X_{max} - X_{min}}
     $$

2. **Standard Scaler**

   * Standardizes features by removing the mean and scaling to unit variance.
   * Formula:

     $$
     X' = \frac{X - \mu}{\sigma}
     $$

   where $\mu$ = mean and $\sigma$ = standard deviation.

### 1. Why imbalance matters

The *Pima Indians Diabetes dataset* has around *500 non-diabetic (0)* and *268 diabetic (1)* cases → a *65:35 imbalance*.

* This is not extreme, but enough for some models to lean toward predicting the majority class (0s).
* In your first confusion matrix, the model leaned *toward 1s* instead, because it optimized for recall of diabetics. That’s not directly because of imbalance, but because of the decision threshold + model training dynamics.

---

### 2. Do you need to balance the dataset?

It depends on your *goal*:

* *If healthcare screening is the goal*:
  You actually want more recall for diabetics (class 1), even if the dataset is imbalanced.
  → In this case, *balancing is not always necessary* because higher recall for 1s is already good.

* *If you want a fair model for both classes (0 and 1)*:
  Then *balancing or using class weights* can help the model treat both classes equally.
  → For example: using class_weight='balanced' in scikit-learn.

---

### 3. Safer Approach in Healthcare

Instead of resampling the dataset, often it’s better to:

* Keep the dataset as is.
* Use *class weights* or *threshold tuning* to favor recall for diabetics.
* Evaluate with *Recall (Sensitivity)* and *Precision* for class 1, not just accuracy.

---

*Conclusion*:
You don’t strictly need to balance the diabetes dataset. Since in healthcare *recall for class 1 (diabetic)* is more critical, your current model is already making the right trade-off (catching more diabetics even at the cost of false alarms).



This matrix shows the performance of a binary classifier (predicting 1 vs 0).

```
                Predicted
              |   1   |   0
Actual   1    |  85   |  15
        0     |  25   |  29
```

### Meaning of the numbers:

* **True Positives (TP) = 85**
  → Model predicted **1** and the actual was **1**.
  (Correct positive predictions)

* **False Negatives (FN) = 15**
  → Model predicted **0** but the actual was **1**.
  (Missed positives)

* **False Positives (FP) = 25**
  → Model predicted **1** but the actual was **0**.
  (Wrongly predicted positive)

* **True Negatives (TN) = 29**
  → Model predicted **0** and the actual was **0**.
  (Correct negative predictions)

---

### Metrics we can calculate:

1. **Accuracy** = (TP + TN) / Total
   \= (85 + 29) / (85 + 15 + 25 + 29)
   \= 114 / 154 ≈ **74%**

2. **Precision (Positive Predictive Value)** = TP / (TP + FP)
   \= 85 / (85 + 25) = 85 / 110 ≈ **77%**

3. **Recall (Sensitivity, True Positive Rate)** = TP / (TP + FN)
   \= 85 / (85 + 15) = 85 / 100 = **85%**

4. **Specificity (True Negative Rate)** = TN / (TN + FP)
   \= 29 / (29 + 25) = 29 / 54 ≈ **54%**

5. **F1-Score** = 2 × (Precision × Recall) / (Precision + Recall)
   ≈ 2 × (0.77 × 0.85) / (0.77 + 0.85) ≈ **81%**

---
### Interpretation

- The model is **good at identifying positives (1s)** → Recall = 85%
- But it struggles more with identifying negatives (0s) → Specificity = 54%
- Overall Accuracy = 74%
- Precision = 77%, meaning most predicted positives are correct

* Class 0 = Non-diabetic
* Class 1 = Diabetic

**Medical perspective:**  

The model is effective at detecting diabetic cases (1s) with a high recall (85%).  
It has lower specificity (54%) for non-diabetic cases (0s), meaning it predicts more false positives.  

This trade-off is acceptable in healthcare datasets because:  
- High recall reduces the chance of missing diabetic patients (false negatives).  
- False positives only require additional testing, which is less risky than missing a diagnosis.  
- Overall, prioritizing recall over specificity is important in medical screening.



````python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Confusion Matrix values
TP, FN, FP, TN = 85, 15, 25, 29

# Build confusion matrix manually
cm = np.array([[TP, FN],
               [FP, TN]])

print("Confusion Matrix:")
print(cm)

# Metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
specificity = TN / (TN + FP)
f1 = 2 * (precision * recall) / (precision + recall)

print("\nPerformance Metrics:")
print(f"Accuracy     : {accuracy:.2%}")
print(f"Precision    : {precision:.2%}")
print(f"Recall (TPR) : {recall:.2%}")
print(f"Specificity  : {specificity:.2%}")
print(f"F1-Score     : {f1:.2%}")
````
