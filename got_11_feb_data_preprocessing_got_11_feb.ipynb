```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/lovnishverma/datasets/refs/heads/main/iriswithissues.csv")

df
```

```python
df.describe()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# fill missing valuesin sepal_length with mean of the sepal_length column
df['sepal_length'].fillna(df['sepal_length'].mean(), inplace=True)
```

```python
# fill missing valuesin sepal_length with mean of the sepal_length column
df['petal_length'].fillna(df['petal_length'].mean(), inplace=True)
```

```python
df.head()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# check for duplicates values
df.duplicated().sum()
```

```python
# drop duplicates
df.drop_duplicates(inplace=True, keep='first')
```

```python
# check for duplicates values
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.head()
```

```python
# encode species column
from sklearn.preprocessing import LabelEncoder # import kiya
le = LabelEncoder() # inilitize
df['species'] = le.fit_transform(df['species'])
```

```python
df.head()
```

```python
df['species'].value_counts()
```

```python
# slacing (normalization)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_df.head()
```
