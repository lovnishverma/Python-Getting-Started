present data in a pictorial or graphical format. It enables analysts and decision makers to see
analytics presented visually, so they can grasp difficult concepts or identify new patterns.

**Line Plots – Using ax.plot()**

Single plot constructed with Figure and Axes in Llisting 2-35

Listing 2-35. Single line plot using ax.plot()

````python
# 1. Generate sample data
x = np.linspace(0, 20, 1000)   # 1000 evenly spaced values from 0 to 20
y = np.sin(x)                  # sine function

# 2. Create a Figure and Axes
fig = plt.figure(figsize=(8, 4))    # Figure: overall container
ax = fig.add_subplot(1, 1, 1)       # Axes: one subplot (1 row, 1 col, 1st plot)

# 3. Plot data on the Axes
ax.plot(x, y, 'b-', linewidth=2, label='Sample label')

# 4. Customize labels, legend, and grid
ax.set_xlabel('x axis label', fontsize=16)
ax.set_ylabel('y axis label', fontsize=16)
ax.legend(loc='best')   # add legend
ax.grid(True)           # show grid

# 5. Adjust layout and add title/footnote
fig.tight_layout(pad=1, rect=[0, 0, 1, 0.95])
fig.suptitle('Sample Plot Title', fontsize=18)   # main title
fig.text(0.99, 0.01, 'Lovnish Verma', ha='right', va='bottom', fontsize=8)

# 6. Save and display the figure
fig.savefig('filename.png', dpi=125, bbox_inches="tight")
plt.show()
plt.close()
````

Listing 2-36. Multiple line plot on same axis

**Multiple Lines on Different Axis**

Instead of plotting all functions on the same Axes, you can create multiple Axes (subplots) inside one Figure. Each subplot gets its own x- and y-axis.

See Listing 2-37

Listing 2-37. Multiple lines on different axis

````python
# 1. Create a Figure with 3 subplots stacked vertically
# Changing sharex to True will use the same x axis
fig, (ax1, ax2, ax3) = plt.subplots(
    nrows=3, ncols=1, sharex=False, sharey=False, figsize=(8, 6)
)

# 2. Generate data
x = np.linspace(0, 100, 200)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y3[np.abs(y3) > 10] = np.nan   # clip tan spikes for readability

# 3. Plot on different Axes
ax1.plot(x, y1, 'b-', linewidth=2, label='sin')
ax2.plot(x, y2, 'g-', linewidth=2, label='cos')
ax3.plot(x, y3, 'r-', linewidth=2, label='tan')

# 4. Customize each Axes
for ax in (ax1, ax2, ax3):
    ax.grid(True)
    ax.legend(loc='best', prop={'size': 'large'})

# 5. Add a common title and adjust layout
fig.suptitle('A Simple Multi-Axis Line Plot', fontsize=16)
fig.tight_layout(pad=1, rect=[0, 0, 1, 0.95])

# 6. Save and show
fig.savefig('multi_axes_plot.png', dpi=125, bbox_inches="tight")
plt.show()
plt.close()
````

**Control the Line Style and Marker Style**

Matplotlib lets you customize both the line style (solid, dashed, dotted, etc.) and the marker style (symbols like circles, crosses, plus signs).
This example shows how to combine them to make different curves easier to distinguish.

See Listing 2-38.

Listing 2-38. Line style and marker style controls

````python
# 1. Create Figure and Axes
fig, ax = plt.subplots(figsize=(8, 4))

# 2. Line and marker styles
N = 3  # number of lines to plot
styles = ['-', '--', '-.', ':']   # line styles
markers = list('+ox')             # marker styles

# 3. Generate data and plot line by line
x = np.linspace(0, 100, 20)
for i in range(N):
    y = x + (x/5)*i + i   # simple linear variation
    s = styles[i % len(styles)]
    m = markers[i % len(markers)]
    ax.plot(
        x, y,
        marker=m,
        linestyle=s,
        linewidth=2,
        alpha=1,
        label=f'Line {i+1} ({s}{m})'
    )

# 4. Customize plot
ax.grid(True)
ax.legend(loc='best', prop={'size': 'large'})
fig.suptitle('A Simple Line Plot with Styles', fontsize=16)

# 5. Save and show
fig.savefig('line_styles.png', dpi=125, bbox_inches="tight")
plt.show()
plt.close()
````

## Colormaps (`cmap`)

Matplotlib has many built-in **colormaps**, grouped into categories:

| Category                                 | Common Colormaps                                                    | Notes                                   |
| ---------------------------------------- | ------------------------------------------------------------------- | --------------------------------------- |
| **Perceptually Uniform (good for data)** | `viridis`, `plasma`, `inferno`, `magma`, `cividis`                  | Best for scientific visualization       |
| **Sequential** (light → dark)            | `Greys`, `Purples`, `Blues`, `Greens`, `Oranges`, `Reds`            | Show ordered progression                |
| **Sequential (multi-hue)**               | `viridis`, `cubehelix`, `YlGnBu`, `PuBuGn`, `YlOrBr`                | Smooth gradients                        |
| **Diverging** (two extremes)             | `PiYG`, `PRGn`, `BrBG`, `RdBu`, `RdGy`, `RdYlBu`, `RdYlGn`          | Best for highlighting differences       |
| **Cyclic**                               | `twilight`, `twilight_shifted`, `hsv`                               | Good for circular data (angles, phases) |
| **Qualitative (categorical)**            | `tab10`, `tab20`, `Set1`, `Set2`, `Set3`, `Pastel1`, `Paired`       | Distinct colors for categories          |
| **Miscellaneous**                        | `flag`, `prism`, `ocean`, `terrain`, `gist_earth`, `rainbow`, `jet` | Often flashy, less recommended for data |

■ Note All color maps can be reversed by appending _r. For instance, gray_r is the
reverse of gray

---

### Example: Applying a colormap

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)
colors = y  # color by value

fig, ax = plt.subplots()
sc = ax.scatter(x, y, c=colors, cmap='viridis')  # apply colormap
fig.colorbar(sc, ax=ax, label="Value")  # add colorbar
plt.show()
```

* This will show points colored by `y` using the **Viridis** colormap.

**Bar Plots – using ax.bar() and ax.barh()**

See Listing 2-39

Listing 2-39. Bar plots using ax.bar() and ax.barh()

````python
# 1. Get the data
N = 4
labels = list('ABCD')
data = np.array(range(N)) + np.random.rand(N)  # add some randomness

# 2. Create Figure and Axes
fig, ax = plt.subplots(figsize=(8, 3.5))

# 3. Bar settings
width = 0.5
tickLocations = np.arange(N)
rectLocations = tickLocations - (width / 2.0)

# 4. Draw vertical bars
ax.bar(
    rectLocations, data, width,
    color='lightblue',              # fill color
    edgecolor='#1f10ed',            # border color (blue-purple)
    linewidth=2.0                   # border thickness
)

# 5. Customize axes
ax.set_xticks(ticks=tickLocations)
ax.set_xticklabels(labels)
ax.set_xlim(min(tickLocations) - 0.6, max(tickLocations) + 0.6)

ax.set_yticks(range(N + 1))   # ticks from 0 to N
ax.set_ylim((0, N))
ax.yaxis.grid(True)

ax.set_ylabel('y axis label', fontsize=10)
ax.set_xlabel('x axis label', fontsize=10)

# 6. Add title and save
fig.suptitle("Vertical Bar Plot", fontsize=14)
fig.tight_layout(pad=2)
fig.savefig('bar_plot.png', dpi=125, bbox_inches="tight")
plt.show()
plt.close()
````

**Horizontal Bar Charts**

Just as tick placement needs to be managed with vertical bars, so it is with horizontal
bars, which are above the y-tick mark as shown in Listing 2-40.


Listing 2-40. Horizontal bar charts

````python
# 1. Get the data
N = 4
labels = list('ABCD')
data = np.array(range(N)) + np.random.rand(N)  # 0,1,2,3 + noise

# 2. Create Figure and Axes
fig, ax = plt.subplots(figsize=(8, 3.5))

# 3. Bar settings
height = 0.5
tickLocations = np.arange(N)
rectLocations = tickLocations - (height / 2.0)

# 4. Draw horizontal bars
ax.barh(
    rectLocations, data, height,
    color='lightblue',      # fill color
    edgecolor='darkblue',   # border color
    linewidth=2.0
)

# 5. Customize axes
ax.set_yticks(ticks=tickLocations)
ax.set_yticklabels(labels)
ax.set_ylim(min(tickLocations) - 0.6, max(tickLocations) + 0.6)

ax.xaxis.grid(True, linestyle="--", alpha=0.7)

ax.set_xlabel('x axis label', fontsize=10)
ax.set_ylabel('y axis label', fontsize=10)

# 6. Add title and save
fig.suptitle("Horizontal Bar Plot", fontsize=14)
fig.tight_layout(pad=2)
fig.savefig('barh_plot.png', dpi=125, bbox_inches="tight")

plt.show()
plt.close()
````

**Side-by-Side Bar Chart**

See Listing 2-41

Listing 2-41. Side-by-side bar chart

````python
# 1. Generate sample data
pre  = np.array([19,  6, 11,  9])
post = np.array([15, 11,  9,  8])
labels = ['Survey ' + x for x in list('ABCD')]

# 2. Create Figure and Axes
fig, ax = plt.subplots(figsize=(8, 3.5))

# 3. Bar settings
width = 0.4
xlocs = np.arange(len(pre))   # positions on x-axis

# 4. Draw bars
ax.bar(xlocs - width/2, pre,  width,
       color='green', edgecolor='black', label='Pre')
ax.bar(xlocs + width/2, post, width,
       color='#1f10ed', edgecolor='black', label='Post')

# 5. Customize axes
ax.set_xticks(ticks=xlocs)
ax.set_xticklabels(labels)
ax.yaxis.grid(True, linestyle="--", alpha=0.7)

ax.set_ylabel('Count', fontsize=10)
ax.set_xlabel('Survey Group', fontsize=10)

# 6. Add legend and title
ax.legend(loc='best')
fig.suptitle('Side-by-Side Bar Chart', fontsize=14)

# 7. Save and show
fig.tight_layout(pad=1)
fig.savefig('side_by_side_bars.png', dpi=125, bbox_inches="tight")

plt.show()
plt.close()
````

**Stacked Bar Example Code**

See Listing 2-42.

Listing 2-42. Stacked Bar Charts (ax.bar with bottom=)

````python
# 1. Generate sample data
pre  = np.array([19,  6, 11,  9])
post = np.array([15, 11,  9,  8])
labels = ['Survey ' + x for x in list('ABCD')]

# 2. Create Figure and Axes
fig, ax = plt.subplots(figsize=(8, 3.5))

# 3. Bar settings
width = 0.4
xlocs = np.arange(len(pre))  # positions on x-axis

# 4. Draw stacked bars
ax.bar(xlocs, pre,  width,
       color='grey', edgecolor='black', label='Pre')
ax.bar(xlocs, post, width,
       color='cyan', edgecolor='black', label='Post',
       bottom=pre)   # stack on top of pre

# 5. Customize axes
ax.set_xticks(ticks=xlocs)
ax.set_xticklabels(labels)
ax.yaxis.grid(True, linestyle="--", alpha=0.7)

ax.set_ylabel('Count', fontsize=10)
ax.set_xlabel('Survey Group', fontsize=10)

# 6. Add legend and title
ax.legend(loc='best')
fig.suptitle('Stacked Bar Chart', fontsize=14)

# 7. Save and show
fig.tight_layout(pad=1)
fig.savefig('stacked_bar.png', dpi=125, bbox_inches="tight")

plt.show()
plt.close()
````

Pie Chart – Using ax.pie()

See Listing 2-43.

Listing 2-43. Pie Chart (ax.pie)

````python
# 1. Generate sample data
data = np.array([15, 8, 4])
labels = ['Feature Engineering', 'Model Tuning', 'Model Building']
explode = (0, 0.1, 0)  # "explode" the 2nd slice
colors = ['cyan', 'tan', 'wheat']

# 2. Create Figure and Axes
fig, ax = plt.subplots(figsize=(8, 3.5))

# 3. Draw pie chart
ax.pie(
    data,
    labels=labels,
    explode=explode,       # offset slice
    autopct='%1.1f%%',     # percentage format
    startangle=270,        # rotate start angle
    colors=colors,
    shadow=True            # optional: add drop shadow
)

# 4. Keep it circular
ax.axis('equal')

# 5. Add title and save
fig.suptitle("ML Pie", fontsize=14)
fig.savefig('pie_chart.png', dpi=125, bbox_inches="tight")

plt.show()
plt.close()
````

**Example Code for Grid Creation**

See Listing 2-44.

Listing 2-44. Grid Creation with fig.add_subplot()

````python
# 1. Create Figure
fig = plt.figure(figsize=(8, 4))

# Add label to the whole figure
fig.text(
    x=0.01, y=0.01,
    s='Figure',
    color='#888888',
    ha='left', va='bottom',
    fontsize=20
)

# 2. Create a 2x2 grid of subplots
for i in range(4):
    # (nrows=2, ncols=2, subplot number)
    ax = fig.add_subplot(2, 2, i + 1)

    # Add text inside each subplot
    ax.text(
        x=0.01, y=0.01,
        s='Subplot 2×2 ' + str(i + 1),
        color='red',
        ha='left', va='bottom',
        fontsize=14
    )

    # Remove ticks
    ax.set_xticks([])
    ax.set_yticks([])

# 3. Add main title and save
fig.suptitle('Subplots', fontsize=16)
fig.tight_layout(pad=2)
fig.savefig('grid_creation.png', dpi=125, bbox_inches="tight")

plt.show()
plt.close()
````

**Plotting – Defaults**

which we call rc settings or rcparameters. You can control the defaults of almost every
property in matplotlib such as figure size and dpi, line width, color and style, axes, axis
and grid properties, text and font properties, and so on. The location of the configuration
file can be found using the code in Listing 2-45 so that you can edit it if required.
