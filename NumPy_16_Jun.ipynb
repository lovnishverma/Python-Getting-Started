**Array Indexing & Slicing**

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr[0])      # First element
print(arr[-1])     # Last element
print(arr[1:4])    # Slicing (from index 1 to 3)
print(arr[::2])    # Every second element

print(arr[::2]) # start:stop:jump(step)
```

```python
arr2d = np.array([[1, 2, 3, 4], [4, 5, 6, 7], [8,9,10,11]])

print(arr2d )
```

```python
print(arr2d[1,1:3])
```

```python
print(arr2d[:,: -1])  # Row, Column
```

```python
print(arr2d[0, 1])  # Row 0, Column 1
print(arr2d[:, 1])  # All rows, Column 1
print(arr2d[1, :])  # Row 1, all columns
```

Arithmetic Operations

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)  # Element-wise addition
print(a - b)  # Element-wise subtraction
print(a * b)  # Element-wise multiplication
print(a / b)  # Element-wise division
print(a ** 2) # Square of each element
```

Aggregations

```python
arr = np.array([1, 2, 3, 4, 5])

print(np.sum(arr))  # Sum of elements
print(np.mean(arr)) # Mean (average)
print(np.max(arr))  # Maximum element
print(np.min(arr))  # Minimum element
print(np.std(arr))  # Standard deviation
```

 Reshaping & Transposing

```python
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped = arr.reshape(2, 3)
print(reshaped)

print(type(reshaped))

print(reshaped.shape)

print(reshaped.ndim)
```

```python
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2d.T)  # Transpose of matrix
```

```python
arr1 = np.array([1, 2, 3, 4, 5, 6])
arr2 = np.array([7, 8, 9, 10, 11, 12])
concatenated = np.concatenate((arr1, arr2))
print(concatenated)
```

```python
arr1 = np.array([1, 2, 3, 4, 5, 6])
arr2 = np.array([7, 8, 9, 10, 11, 12])

# Horizontal stacking

h_stack = np.hstack((arr1, arr2))
print(h_stack)

print("---------------------------------------")

# Vertical stacking
v_stack = np.vstack((arr1, arr2))
print(v_stack)
```

**Useful Numpy fuctions**

```python
arr = np.array([1,2,3,4,4,5,6,6])

unique = np.unique(arr)
print(unique)
```

```python
arr = np.array([1,2,7,3,4,4,5,6,6])

sort = np.sort(arr)
print(sort)
```

```python
arr = np.array([1,2,7,3,8,4,4,5,6,6])

where_vals = np.where(arr == 6)

print(where_vals)
```

Mini Project example: Normalization

```python
data = np.array([50, 100, 150, 200])

normalized = (data - np.min(data)) / (np.max(data) - np.min(data))

print(normalized)
```
