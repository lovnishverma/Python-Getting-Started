## Create Flask Application

````python
app_code = """
from flask import Flask, request, render_template_string
import requests
from bs4 import BeautifulSoup

app = Flask(__name__)

HTML_TEMPLATE = \"\"\"
<!DOCTYPE html>
<html>
<head>
    <title>Web Scraper</title>

    <style>
        body {
            font-family: Arial;
            margin: 40px;
        }

        input, button {
            padding: 10px;
            margin: 5px;
        }
    </style>

</head>

<body>

<h1>Simple Web Scraper</h1>

<form method="POST">

    <input
        type="text"
        name="url"
        placeholder="Enter URL"
        required
    >

    <input
        type="text"
        name="tag"
        placeholder="Enter HTML Tag"
        required
    >

    <button type="submit">
        Scrape
    </button>

</form>

{% if results %}

<h2>Results</h2>

<ul>
{% for item in results %}
    <li>{{ item }}</li>
{% endfor %}
</ul>

{% endif %}

</body>
</html>
\"\"\"

@app.route("/", methods=["GET", "POST"])
def home():

    results = []

    if request.method == "POST":

        url = request.form.get("url")
        tag = request.form.get("tag")

        headers = {
            "User-Agent": "Mozilla/5.0"
        }

        try:

            response = requests.get(
                url,
                headers=headers,
                timeout=10
            )

            response.raise_for_status()

            soup = BeautifulSoup(
                response.text,
                "html.parser"
            )

            elements = soup.find_all(tag)

            results = [
                e.get_text(strip=True)
                for e in elements
            ]

        except Exception as e:
            results = [f"Error: {str(e)}"]

    return render_template_string(
        HTML_TEMPLATE,
        results=results
    )

if __name__ == "__main__":
    app.run(
        host="0.0.0.0",
        port=5000
    )
"""

with open("app.py", "w", encoding="utf-8") as f:
    f.write(app_code)

print("app.py created successfully")
````
