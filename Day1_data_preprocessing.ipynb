# Two core DS in Pandas

* Series

* Dataframe

```python
# Series in pandas
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)

series
```

```python
# Dataframe in Pandas
data = {
    'Name': ['Japneet', 'Akshada', 'Babita'],
    'Age': [26, 23, 50],
    'City': ['Delhi', 'Shimla', 'Ropar'],
    'Salary': [45000,35000,60000]}

df = pd.DataFrame(data)
df
```

```python
# Custom indexing in datafram
data = {
    'Name': ['Japneet', 'Akshada', 'Babita'],
    'Age': [26, 23, 50],
    'City': ['Delhi', 'Shimla', 'Ropar'],
    'Salary': [45000,35000,60000]}

df = pd.DataFrame(data, index=['A','B','C'])

df
```

```python
# Dataframe in Pandas
data = {
    'Name': ['Japneet', 'Akshada', 'Babita'],
    'Age': [26, 23, 50],
    'City': ['Delhi', 'Shimla', 'Ropar'],
    'Salary': [45000,35000,60000]}

df = pd.DataFrame(data)
df
```
