**Matplotlib**, is the most popular data visualization library in Python â€” **clear, visual, and beginner-friendly** explanation.

## ðŸŽ¯ What is Matplotlib?

- **Matplotlib** is a Python **library for creating static, interactive, and animated plots**.
- It's similar to **Excel charts** or **plots in MATLAB**, but for Python.
- It's highly customizable and supports a wide variety of plot types: line, bar, scatter, pie, histogram, box, and many more.

---

## ðŸ§  How It Works: Big Picture

Matplotlib has two main interfaces:

| Interface      | Description                                | Example                        |
|----------------|--------------------------------------------|--------------------------------|
| `pyplot`       | Simple, MATLAB-like interface (easy to use) | `plt.plot(...)`, `plt.show()` |
| `Object-Oriented` | Advanced control, figure & axes objects   | `fig, ax = plt.subplots()`    |

---

## âœ¨ Basic Syntax (Pyplot Style)

```python
import matplotlib.pyplot as plt

# X and Y values
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)               # Line plot
plt.title("My Line Graph")   # Title
plt.xlabel("X Axis")         # X label
plt.ylabel("Y Axis")         # Y label
plt.grid(True)               # Show grid
plt.show()                   # Display the plot
```

---

## ðŸ“¦ Common Matplotlib Charts

| Type         | Function        | Example                              |
|--------------|------------------|--------------------------------------|
| Line         | `plt.plot()`     | `plt.plot(x, y)`                     |
| Bar          | `plt.bar()`      | `plt.bar(x, height)`                 |

| Histogram    | `plt.hist()`     | `plt.hist(data, bins=5)`             |
| Pie          | `plt.pie()`      | `plt.pie(data)`                      |
| Box Plot     | `plt.boxplot()`  | `plt.boxplot(data)`                  |
| Area (stack) | `plt.stackplot()`| `plt.stackplot(x, y1, y2, ...)`      |

## ðŸ”§ Customization Examples

```python

```

| Customization     | Use                            |
|-------------------|--------------------------------|
| `color='red'`     | Line color                     |
| `linestyle='--'`  | Dashed line                    |
| `marker='o'`      | Dots on data points            |
| `linewidth=2`     | Thickness of line              |
| `legend([...])`   | Add legend                     |

---

## ðŸ§± Figure, Axes, and Subplots (Object-Oriented)

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Line Chart")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

| Component | Meaning                                       |
|-----------|-----------------------------------------------|
| `Figure`  | The entire canvas (can contain multiple plots)|
| `Axes`    | A single plot area                            |
| `Subplot` | Multiple plots in one figure                  |

## ðŸ“¥ Saving Plots

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("https://raw.githubusercontent.com/lovnishverma/datasets/refs/heads/main/matplotlibcsv.csv")
# print(df)

df.head()       # View first 5 rows
# df["Month"]      # Access a column
# df.iloc[0]      # Access first row
# df.describe()   # Summary statistics
````

````python
#LINE CHART: Sales over Months
plt.plot(df["Month"], df["Sales"], marker='o', color='blue')
plt.title("Line Chart: Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.savefig("lineplot.png")
plt.show()
````

````python
#BAR CHART: Profit over Months
plt.bar(df["Month"], df["Profit"], color='green')
plt.title("Bar Chart: Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()
````

````python
#HORIZONTAL BAR CHART: Expenses
plt.barh(df["Month"], df["Expenses"], color='orange')
plt.title("Horizontal Bar Chart: Monthly Expenses")
plt.xlabel("Expenses")
plt.ylabel("Month")
plt.show()
````

````python
#HISTOGRAM: Units Sold distribution
plt.hist(df["Units_Sold"], bins=5, color='purple', edgecolor='black')
plt.title("Histogram: Units Sold")
plt.xlabel("Units Sold")
plt.ylabel("Frequency")
plt.show()
````

````python
#SCATTER PLOT: Sales vs Profit
plt.scatter(df["Sales"], df["Profit"], color='red')
plt.title("Scatter Plot: Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
````

````python
#PIE CHART: Sales distribution by month
plt.pie(df["Sales"], labels=df["Month"], autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart: Sales Share")
plt.show()
````

````python
#BOX PLOT: Profit distribution
print(df.head(5))
plt.boxplot(df["Profit"], patch_artist=True, boxprops=dict(facecolor="skyblue"))
plt.title("Box Plot: Profit")
plt.ylabel("Profit")
plt.show()
````

````python
#AREA CHART: Sales and Expenses over time (stacked)
plt.stackplot(df["Month"], df["Sales"], df["Expenses"], labels=["Sales", "Expenses"], colors=["#4CAF50", "#F44336"])
plt.title("Area Chart: Sales and Expenses")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend(loc="upper left")
plt.show()
````

## ðŸ“Š When to Use What in Seaborn

| Plot Type       | Function           | Best For                       |
|-----------------|--------------------|--------------------------------|
| `barplot()`     | Bar chart          | Category vs Value              |
| `lineplot()`    | Line chart         | Trends                         |
| `scatterplot()` | Dots               | Correlations                   |
| `histplot()`    | Histogram + KDE    | Distributions                  |
| `boxplot()`     | Box plot           | Outliers + spread              |
| `heatmap()`     | Colored matrix     | Correlation matrix, tables     |

---

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("https://raw.githubusercontent.com/lovnishverma/datasets/refs/heads/main/matplotlibcsv.csv")

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Bar Plot: Month vs Profit
sns.barplot(x="Month", y="Profit", hue="Month", data=df, palette="crest", legend=False)
plt.title("Seaborn Barplot: Monthly Profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Line Plot: Sales Trend
sns.lineplot(x="Month", y="Sales", data=df, marker="o", color="orange")
plt.title("Seaborn Lineplot: Monthly Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter Plot: Sales vs Profit
sns.scatterplot(x="Sales", y="Profit", data=df, s=100, color="red")
plt.title("Seaborn Scatterplot: Sales vs Profit")
plt.tight_layout()
plt.show()

# Histogram: Distribution of Units Sold
sns.histplot(df["Units_Sold"], bins=5, kde=True, color="purple")
plt.title("Seaborn Histogram: Units Sold")
plt.tight_layout()
plt.show()

# Box Plot: Profit distribution
sns.boxplot(y="Profit", data=df, color="skyblue")
plt.title("Seaborn Boxplot: Profit")
plt.tight_layout()
plt.show()
````
