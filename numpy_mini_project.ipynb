```python
import numpy as np

print(np.__version__)
```

```python
students = np.array(["Aman", "Bhumika", "Chaman", "Dinesh", "Ekta", "Faruk"])

marks = np.array([85, 95, 66, 38, 55, 71])
```

```python
# find marks for each students
for i in range(len(students)):
    print(students[i], marks[i])
```

```python
# Avegare marks
average_marks = np.mean(marks)
print(average_marks)
```

```python
# minimum marks
min_marks = np.min(marks)
print(min_marks)


min_index = np.argmin(marks)
print(min_index)

print(students[min_index])
```

```python
# maximun marks
max_marks = np.max(marks)
print(max_marks)

max_index = np.argmax(marks)
print(max_index)

print(students[max_index])
```

```python
# check passed students whose score is >= 40

passed_students = marks >= 40
print(passed_students)
```

```python
# Check passing students names with score
for i in range(len(students)):
    if marks[i] >= 40:
        print(students[i], marks[i])
```

```python
# check failed students whose score is < 40
failed_students = marks < 40
print(failed_students)
```

```python
# Check failing students names with score
for i in range(len(students)):
    if marks[i] < 40:
        print(students[i], marks[i])
```
