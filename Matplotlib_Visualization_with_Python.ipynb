**2. Creating a Simple Line Plot**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Sine function

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**Explanation**:

`np.linspace(0, 10, 100)`: Creates 100 evenly spaced points between 0 and 10.

`plt.plot(x, y)`: Plots y = sin(x).

`plt.title(), plt.xlabel(), plt.ylabel()`: Adds labels.

`plt.show()`: Displays the plot.

**3. Customizing Line Styles & Markers**

```python
plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='o', markersize=5)
plt.show()
```

**Customization Options**:

`color='red'`: Sets line color.

`linestyle='--'`: Dashed line.

`linewidth=2`: Line thickness.

`marker='o'`: Circular markers.

`markersize=5`: Marker size.

**4. Multiple Line Plots**

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="Sine", color="blue")
plt.plot(x, y2, label="Cosine", color="green")

plt.legend()  # Show legend
plt.title("Sine and Cosine Functions")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**5. Scatter Plot**

```python
x = np.random.rand(50)  # 50 random points
y = np.random.rand(50)

plt.scatter(x, y, color='purple', marker='x')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**6. Bar Chart**

```python
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 30]

plt.bar(categories, values, color='orange')
plt.title("Bar Chart Example")
plt.show()
```

**7. Histogram**

```python
data = np.random.randn(1000)  # 1000 random numbers from normal distribution

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.show()
```

**8. Pie Chart**

```python
labels = ["Apples", "Bananas", "Cherries", "Dates"]
sizes = [30, 40, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['red', 'yellow', 'pink', 'brown'])
plt.title("Pie Chart Example")
plt.show()
```

**9. Subplots (Multiple Plots in One Figure)**

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))  # 2x2 grid of subplots

x = np.linspace(0, 10, 100)
y = np.sin(x)

axs[0, 0].plot(x, y, 'r')  # Top-left
axs[0, 1].scatter(x, np.cos(x), color='blue')  # Top-right
axs[1, 0].bar(["A", "B", "C"], [5, 7, 3])  # Bottom-left
axs[1, 1].hist(np.random.randn(500), bins=20)  # Bottom-right

plt.tight_layout()  # Adjust layout
plt.show()
```

1️⃣ Subplots
Subplots help you show multiple plots in a single figure.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

# Create subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot on each subplot
axes[0, 0].plot(x, np.sin(x), 'r')  # Red sine wave
axes[0, 0].set_title('Sine Wave')

axes[0, 1].plot(x, np.cos(x), 'g')  # Green cosine wave
axes[0, 1].set_title('Cosine Wave')

axes[1, 0].scatter(x, np.random.rand(100), color='blue')
axes[1, 0].set_title('Random Scatter')

axes[1, 1].hist(np.random.randn(1000), bins=30, color='purple')
axes[1, 1].set_title('Histogram')

plt.tight_layout()  # Adjust layout to avoid overlapping
plt.show()
```

2️⃣ Customizing Plots (Colors, Markers, and Styles)
You can customize your plots using different colors, markers, and line styles.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Customize line style, color, and marker
plt.plot(x, y, linestyle='--', color='orange', marker='o', markersize=5, label='Sine')
plt.title('Customized Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)  # Add grid lines
plt.show()
```

**Line Styles**: '-', '--', '-.', ':'

**Colors**: 'r', 'g', 'b', 'orange', etc.

**Markers**: 'o', 's' (square), 'x', '*', 'D' (diamond), etc.

3️⃣ Saving Plots as Images
You can save plots as PNG, PDF, SVG, and more using savefig().

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Plot to be Saved')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Save the plot as a PNG image
plt.savefig('saved_plot.png', dpi=300, bbox_inches='tight')  # High resolution
plt.show()
```

4️⃣ Animated Plots
Matplotlib supports animations using FuncAnimation from matplotlib.animation. Here’s a simple animation example:

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()
line, = ax.plot(x, np.sin(x))

# Function to update the plot for each frame
def update(frame):
    line.set_ydata(np.sin(x + frame / 10))  # Change sine wave phase
    return line,

ani = FuncAnimation(fig, update, frames=100, interval=50)  # 100 frames, 50 ms interval
plt.show()
```

FuncAnimation Parameters:

frames: Total number of frames

interval: Delay between frames (in milliseconds)

update: Function to update the plot

You can save the animation as a video using ani.save('animation.mp4').
