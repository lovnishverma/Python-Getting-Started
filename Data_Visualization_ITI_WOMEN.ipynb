```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/lovnishverma/datasets/refs/heads/main/datacc.csv") #  GitHub raw link for code reproducibiility
```

# Bar Chart

```python
plt.bar(df['State'], df['Crime_Rate'], color=['g','r', 'purple','orange'])
plt.xlabel('Statessss')
plt.ylabel('Crime Rate')
plt.title('State vs Crime Rat latest wala')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Crime_Rate', y='State', data=df, palette='viridis', hue='State', legend=False)
plt.title('Amazing State vs Crime Rate Plot')
plt.xlabel('Crime Rate')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

# Line plot

```python
plt.plot(df['State'], df['Crime_Rate'], marker='o', color='red', linestyle='--', label='Crime Rate')
plt.xlabel('States')
plt.ylabel('Crime Rate')
plt.title("State vs Crime Rate")
plt.grid()
plt.savefig('plotmerapyara.png', dpi=1200) # always save before show()
plt.legend()
plt.show()

```

```python
a = ['a','b']
b = [10,50]

plt.bar(a, b, color=['g','r', 'purple','orange'])
plt.xlabel('Statessss')
plt.ylabel('Crime Rate')
plt.title('State vs Crime Rat latest wala')
plt.show()
```

# Scatter Plot()

```python
plt.scatter(df['State'], df['Crime_Rate'])
plt.xlabel('States')
plt.ylabel('Crime Rate')
plt.title("State vs Crime Rate")
plt.show()

```

# Pie Chart

```python
#pie chart
plt.pie(df['Crime_Rate'], labels=df['State'], autopct='%1.1f%%')
plt.title('State vs Crime Rate')
plt.show()
```
