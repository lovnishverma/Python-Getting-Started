# Arithmetic Operations

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
print(a + b)
print(a - b)
print(a * b)
print(a / b) # divide
print(a ** 2) # square of each element
```

# Stacking and Concetenation

**Horizontal Stacking**

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.hstack((a,b)))
```

**Verical Stacking**

```python
a = np.array([[1,2], [3,4]])
b = np.array([[5,6], [7,8]])

print(np.vstack((a,b)))
```

# Useful Numpy fuctions

```python
arr = np.array([1,2,3,4,4,5,6,6])

unique = np.unique(arr)

print(unique)
```

```python
arr = np.array([1,2,7,3,8,4,4,5,6,6])

sorted = np.sort(arr)

print(sorted)
```

```python
arr = np.array([1,2,7,3,8,4,4,5,6,6])

where_vals = np.where(arr == 5)

print(where_vals)
```

# Mini Project example: Normalization

```python
data = np.array([50, 100, 150, 200])

normalized = (data - np.min(data)) / (np.max(data) - np.min(data))

print(normalized)
```
